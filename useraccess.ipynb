```python
class DateTimeEncoder(json.JSONEncoder):
    def default(self, obj):
        if isinstance(obj, dt.date):
            return obj.isoformat()
        return json.JSONEncoder.default(self, obj)
```

```python
def add_get_user():
    while True:
        fname = f_name("Enter your first name: ")
        lname = l_name("Enter your last name: ")
        email = u_email("Enter your email: ")
        password = u_password("Enter your password: ")
        con_pass = u_cpassword("Confirm password: ")
        phone = u_phone("Enter your phone: ")

        while password != con_pass:
            print("Passwords do not match. Please enter again.")
            password = u_password("Enter your password: ")
            con_pass = u_cpassword("Confirm password: ")

        id = generate_id()
        user_data = {
            "id": id,
            "First Name": fname,
            "Last Name": lname,
            "Email": email,
            "Password": password,
            "Phone": phone
        }
        save_new_data(user_data)
        print("User registered successfully.")
        break
```

```python
def log_in():
    while True:
        email = u_email("Enter your email: ")
        password = u_password("Enter your password: ")
        users_data = get_all_data()
        found = False  # Variable to keep track if the user is found

        for user in users_data:
            if user['Email'] == email and user['Password'] == password:
                print("------------You are logged in-------------------")
                found = True  # Set found to True if the user is found
                break

        if found:
            break  # Break the while loop if the user is found
        else:
            print("----------Please enter your email and password again--------------")
```

```python
def update_pr_u():
    old_data_p = get_all_project()
    old_data_p_u = get_all_pd()
    id_to_find = askforid("Please enter User id: ")
    for u_p in old_data_p_u:
        if u_p['User Id'] == id_to_find:
            print("-----Found User-----")
            print(u_p)
            for pro in old_data_p:
                if u_p['Project Id'] == pro['id']:
                    ntitle = p_title("Enter the edit Title of projects ")
                    ndetails = p_details("Enter edit Details of project ")
                    ntarget = p_target("Enter edit Target of projects ")
                    nstart_date = S_date("Enter edit Start Date of project ")
                    nend_date = E_date("Enter edit End Date of project ")

                    # Update project data
                    pro['Title'] = ntitle
                    pro['Details'] = ndetails
                    pro['Target'] = ntarget
                    pro['Start Date'] = nstart_date
                    pro['End Date'] = nend_date
                    
                    # Save updated data to file
                    valid_data = json.dumps(old_data_p, indent=2, cls=DateTimeEncoder)
                    try:
                        with open("projects.json", "w") as file:
                            file.write(valid_data)
                        print("Updated Successfully")
                        return True, pro  # Assuming you want to return the last modified project
                    except Exception as e:
                        print(f"An error occurred while saving data: {e}")
                        return False
            else:
                print("---------Project ID not found for this User--------------")
                return False
    else:
        print("---------User ID not found--------------")
        return False
```

```python
def delete_pro():
    old_data_p = get_all_project()
    old_data_p_u = get_all_pd()
    id_to_find = askforid("Please enter User id: ")
    for u_p in old_data_p_u:
        if u_p['User Id'] == id_to_find:
            print("-----Found User-----")
            print(u_p)
            for pro in old_data_p:
                if u_p['Project Id'] == pro['id']:
                    old_data_p.remove(pro)  # Remove the project from the list
                    print("Project removed successfully.")
                    with open('projects.json', 'w') as file:
                        json.dump(old_data_p, file, indent=2, cls=DateTimeEncoder)
                    break  # Exit the loop once project is removed
            else:
                print("Project not found for this user.")
            break  # Exit the loop once user is found
    else:
        print("Project not found.")

# Assuming 'projects.json' is the JSON file containing project data
```

```python
def mainmenu():
    while True:
        print("----------------------------------")
        choice = input("""Please enter 
                            1 for new user, 
                            2 for log in ,
                            3 for exit: """)
        if choice.isdigit():
            choice = int(choice)  # Convert input to integer for comparison
            
            if choice == 1:
                add_get_user()
                continue
            elif choice == 2:
                log_in()
                kind_find = input("""Enter 
                                    1 For search by Id
                                    2 for Start Date
                                    3 for End Date
                                    4 for add New Project
                                    5 for update project 
                                    6 for delete project
                                    7 for list projects
                                    8 for add new access
                                    9 for delete access
                                    """)
                if kind_find.isdigit():
                    kind_find = int(kind_find)  # Convert input to integer for comparison
                    if kind_find == 1:
                        find_pro()
                    elif kind_find == 2:
                        get_pro_D_S()
                    elif kind_find == 3:
                        get_pro_D_E()
                    elif kind_find == 4:
                        new_project()
                    elif kind_find == 5:
                        update_pr_u()
                    elif kind_find == 6:
                        delete_pro()
                    elif kind_find == 7:
                        print(get_all_project())
                    elif kind_find == 8:
                        realtion_p_u()
                    elif kind_find == 9:
                        delete_p_u()
                    else:
                        print("Please enter a valid choice")
                else:
                    print("Please make sure you are entering a numeric input ")
            elif choice == 3:
                exit()
            else:
                print("Please enter a valid choice")
        else:
            print("Please make sure you are entering a numeric input")
```
